```python
import pandas as pd
import chardet as cd
import matplotlib.pyplot as plt


with open(r"../../Datasets+Data+Science+con+Python+-+Pandas/1.3/Info_pais.csv","rb") as file:
    resultado = cd.detect(file.read())
print(resultado)
```

```python
poblacion = pd.read_csv("../../Datasets+Data+Science+con+Python+-+Pandas/1.3/Info_pais.csv" , delimiter=";" , encoding="ISO-8859-1")
```

```python
ordenamiento =poblacion.sort_values("Esperanza de vida",ascending=True)
ordenamiento.head()
```

```python
plt.scatter(ordenamiento["Renta per capita"],ordenamiento["Esperanza de vida"])
plt.title("Renta")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
```

```python
ordenamiento["Poblacion_normalizad"] = ordenamiento["Poblacion"]/(max(ordenamiento["Poblacion"])/10000)
```

```python
ordenamiento
```

```python
plt.scatter(ordenamiento["Renta per capita"],ordenamiento["Esperanza de vida"],s=ordenamiento["Poblacion_normalizad"])
plt.title("Renta")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
figura = plt.gcf()
figura.set_size_inches(14.5,10)
figura.savefig("bb")
```

```python
plt.scatter(ordenamiento["Renta per capita"],ordenamiento["Esperanza de vida"],s=ordenamiento["Poblacion_normalizad"],c=ordenamiento["Poblacion_normalizad"])
plt.title("Renta")
plt.xlabel("Renta per capita")
plt.ylabel("Esperanza de vida")
figura = plt.gcf()
figura.set_size_inches(14.5,10)
for i in range(1,10):
    plt.annotate(ordenamiento["País"][i],(ordenamiento["Renta per capita"][i],ordenamiento["Esperanza de vida"][i]))

plt.yticks(ticks=range(1,120,10))
```
